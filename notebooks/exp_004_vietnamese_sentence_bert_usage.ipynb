```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
visbert_tokenizer = AutoTokenizer.from_pretrained('keepitreal/vietnamese-sbert')
visbert = AutoModel.from_pretrained('keepitreal/vietnamese-sbert') 
```

```python
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings 
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

def visbert_encode(sentences):
    tokens = visbert_tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        outputs = visbert(**tokens)
    embeddings = mean_pooling(outputs, tokens['attention_mask'])
    return embeddings
```

```python
%%time

sentences = [
    'Cô giáo đang ăn kem',
    'Chị gái đang thử món thịt dê',
]

embeddings = visbert_encode(sentences)
print('Embedding data type:', embeddings.dtype)
print('Embedding size:', embeddings.size())
```
